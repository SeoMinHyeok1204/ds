```python
#실습1
import csv
import matplotlib.pyplot as plt
f = open('seoul_10years.csv')
data = csv.reader(f)
next(data)
max = []
min = []

for row in data:
    if (row[-1] != ''):
        if(row[0].split('-')[1] == '12' and row[0].split('-')[2] == '04'):
            max.append(float(row[-1]))
            min.append(float(row[-2]))
plt.plot(max)
plt.plot(min)
plt.show()
```

```python
#실습2
import csv
import matplotlib.pyplot as plt
f = open('seoul_10years.csv')
data = csv.reader(f)
next(data)
jan = []
aug = []

for row in data:
    if (row[-1] != ''):
        if(row[0].split('-')[1] == '01'):
            jan.append(float(row[-1]))
        elif(row[0].split('-')[1] == '08'):
            aug.append(float(row[-1]))
        
plt.hist(jan, bins = 100, color = 'r', label = 'jan')
plt.hist(aug, bins = 100, color = 'b', label = 'aug')
plt.legend(loc=9)
plt.show()
```
